### Label-based slicing with loc[] - second index INCLUSIVE

```python
# view specific rows by index with loc[]

djson.loc[5]
```

```python
# view multiple rows

djson.loc[[2, 4, 7]]
```

```python
# view a range of rows

djson.loc[3:5]
```

```python
# view a range of rows AND a range of columns (first rows, then columns)

djson.loc[0:2, "Phone":"Skills"]
```

```python
# return all the values for a specific column

djson.loc[:, "Phone"]
```

```python
# finding the unique elements in a column

set(djson.loc[:, "Department"])
```

### Position-based slicing with iloc[] - second index EXCLUSIVE

```python
djson.iloc[5]
```

```python
# second column is not included

djson.iloc[2, 2:5]
```

### Using boolean operators

```python
# the or operator '|'

djson[(djson.loc[:,"Salary"] < 50000) | (djson.loc[:,"Salary"] > 56000)]
```

```python
# the and operator '&'

djson[(djson.loc[:,"Salary"] > 50000) & (djson.loc[:,"Department"] == "Sales")]
```

```python
# the not logical operator '~'

djson[~(djson.loc[:, "Department"] == "IT")]
```

### Adding a row

```python
djson = djson.append([{"Address": "11th Address, Miami", 
                      "Department": "IT",
                      "FirstName": "John",
                      "ID": 11,
                      "LastName": "Sykes",
                      "Phone": "00999999",
                      "Salary": "600000",
                      "Skills": "Java",
                      "Badge ID": "0020"},
                     {"Address": "12th Address, Miami", 
                      "Department": "IT",
                      "FirstName": "Steve",
                      "ID": 12,
                      "LastName": "Kane",
                      "Phone": "90999999",
                      "Salary": "650000",
                      "Skills": "Python",
                      "Badge ID": "0021"}], ignore_index = True)
```

```python
# check the additions

djson
```

```python
# updating a column

djson["Badge ID"] = djson["FirstName"] + "2021"
djson
```

```python
# update multiple rows

djson["Skills"] = djson.shape[0] * ["Python, Tableau"]
djson
```

```python
# updating multiple rows based on a condition

djson.loc[djson["Department"] == "IT", "Salary"] = '90000'
djson
```

```python
# update multiple fields on the same row

djson.loc[djson["ID"] == 8, ["Salary", "Phone"]] = ["80000", "777777777"]
djson
```

```python
# change the index

djson.set_index("Badge ID", inplace = True)
djson
```

```python
# delete a row by its label (change is temporary bc "inplace = True" is not being used)

djson.drop("Luke2021")
```
